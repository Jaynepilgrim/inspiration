Благодаря этому задача удержания будет легче в реализации, чем задача привлечения новых пользователей, о которых мы еще ничего не знаем.<br>

# 1. Описание данных

```python
### YOUR CODE HERE ###
# Load in our libraries
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

```

```python
# Load in the dataset
dataset = pd.read_csv('telecom_users.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset  = dataset.drop(['customerID'], axis = 1)
dataset  = dataset.drop(['Unnamed: 0'], axis = 1)
dataset.head()
```

```python
dataset.dtypes
```

```python
for col_name in dataset.columns:
    if(dataset[col_name].dtype == 'object'):
        dataset[col_name]= dataset[col_name].astype('category')
        dataset[col_name] = dataset[col_name].cat.codes
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

Клиентов мужчин и женщин почти поровну, половина клиентов состоят в браке, в среднем клиенты являются клиентами компании 32 месяца, почти у всех клиентов подключена услуга телефонной связи. У большей половины пользовательей подключены несколько телефонных линий. У клиентов в основном подключены дополнительные услуги. Наблюдается сильный отток клиентов из компании. Стоимость услуг в месяц варьируется от 18.25 до 118.75. Большая половина клиентов пользуется безбумажным биллингом.

# 3. Построение моделей для прогнозирования оттока

```python
### YOUR CODE HERE ###
#load in libraries
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
features_all = dataset.drop(['Churn'], axis = 1)
target_all = dataset['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_all, target_all, test_size = 0.25, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn_model = knn.fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
knn_predictions
```

```python
nb = GaussianNB()
nb_model = nb.fit(X_train, y_train)
nb_predictions = nb.predict(X_test)
nb_predictions
```

```python
dtc = DecisionTreeClassifier(random_state = 42)
dtc_model = dtc.fit(X_train, y_train)
```

```python
dtc_predictions = dtc.predict(X_test)
dtc_predictions
```

# 4. Сравнение качества моделей 

```python
### YOUR CODE HERE ###
#Использую метрику точность, вывожу матрицу ошибок.
from sklearn.metrics import confusion_matrix, precision_score
```

```python
conf_mtrx_knn = confusion_matrix(y_test, knn_predictions)
conf_mtrx_knn
```

```python
precision_score_knn = precision_score(y_test, knn_predictions)
print(f'precision_score_knn: {precision_score_knn}')
```

```python
conf_mtrx_nb = confusion_matrix(y_test, nb_predictions)
conf_mtrx_nb
```

```python
precision_score_nb = precision_score(y_test, nb_predictions)
print(f'precision_score_nb: {precision_score_nb}')
```

```python
conf_mtrx_dtc = confusion_matrix(y_test, dtc_predictions)
conf_mtrx_dtc
```

```python
precision_score_dtc = precision_score(y_test, dtc_predictions)
print(f'precision_score_dtc: {precision_score_dtc}')
```

наилучшую точность precision_score_knn: 0.5210355987055016 показывает модель KNN.

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, knn_model.predict_proba(X_test)[:, 1])
```

```python
roc_auc_score(y_test, nb_model.predict_proba(X_test)[:, 1])
```

```python
roc_auc_score(y_test, dtc_model.predict_proba(X_test)[:, 1])
```

```python
from sklearn.metrics import precision_recall_fscore_support, classification_report
```

```python
precision_recall_fscore_support_knn = precision_recall_fscore_support(y_test, knn_predictions)
print(f'precision_recall_fscore_support_knn: {precision_recall_fscore_support_knn}')
```

```python
precision_recall_fscore_support_nb = precision_recall_fscore_support(y_test, nb_predictions)
print(f'precision_recall_fscore_support_nb: {precision_recall_fscore_support_nb}')
```

```python
precision_recall_fscore_support_dtc = precision_recall_fscore_support(y_test, dtc_predictions)
print(f'precision_recall_fscore_support_dtc: {precision_recall_fscore_support_dtc}')
```
